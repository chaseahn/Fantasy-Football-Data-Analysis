```python
df['t_pts'] = (df['Tgt'] + df['TD']*6 + df['Yds']*.1)
```

```python
df['f_ppg'] = (df['t_pts']/df['G']).round(1)
```

```python
"""LAMBDA FUNCTIONS"""

lambda arguments: expression

lambda x: x*x

def square(x):
    return x*x
```

```python
def name_split(name):
    name = name.split('\\')[0]
    return name
df['Unnamed: 1'] = df.apply(lambda row: name_split(row['Unnamed: 1']), axis=1)
```

```python
def name_strip(name):
    name = name.strip('*+\\')
    return name
df['Unnamed: 1'] = df.apply(lambda row: name_strip(row['Unnamed: 1']), axis=1)
```

```python
#df.isna()
df = df.fillna('0')
df.iloc[34]
```

```python
def ctch_strip(name):
    name = float(name.strip('%'))
    return name
df['Ctch%'] = df.apply(lambda row: ctch_strip(row['Ctch%']), axis=1)
```

```python
df.set_index('Unnamed: 1', inplace=True)
```

```python
df.index.rename('Name', inplace=True)
```
